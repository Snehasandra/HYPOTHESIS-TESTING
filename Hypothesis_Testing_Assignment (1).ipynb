#Cutlets

```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
from google.colab import files
uploaded=files.upload()
```

```python
Data=pd.read_csv('Cutlets.csv')
Data
```

```python
Data['Unit A'].skew()
```

```python
import seaborn as sns
sns.distplot(Data['Unit A'])
```

```python
Data['Unit B'].skew()
```

```python
sns.distplot(Data['Unit B'])
```

NULL HYPOTHESIS - The diameter of the cutlet between two units are significant.

ALTERNATE HYPOTHESIS - There is no significant diameter of the cutlet between two units

```python
stats.f_oneway(Data['Unit A'],Data['Unit B'])
```

If the p_value is less than 5% Alternate Hypothesis test is accepted
If it is greater than 5% so Null Hypothesis test is accepted

In this case we get the p_value less than 5% that means Alternate Hypothesis will be accepted

```python
Data=pd.read_csv('LabTAT.csv')
Data
```

NULL HYPOTHESIS - There is no any difference in average TAT among the different laboratories

ALTERNATE HYPOTHESIS -There is difference in average TAT among the difference laboratories

```python
Data['Laboratory 1'].skew()
```

```python
Data['Laboratory 2'].skew()
```

```python
Data['Laboratory 3'].skew()
```

```python
Data['Laboratory 4'].skew()
```

```python
stats.f_oneway(Data['Laboratory 1'],Data['Laboratory 2'],Data['Laboratory 3'],Data['Laboratory 4'])
```

If the significant value is greater than 0.05 the Null Hypothesis is accepted.
If the significant value is less than 0.05 the Alternate Hypothesis is accepted.


Here we get greater than 0.05,so the Null Hypothesis is accepted
